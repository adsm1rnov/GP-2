Удаление столбцов, не представляющих важности в рамках нашего исследования: индексы, ссылки на сайт и изображения, технические индексы, полученные в результате парсинга

```python
del hotels['image']
del hotels['Unnamed: 0']
del hotels['@type']
del hotels['@context']
del hotels['url']
del hotels['hasMap']
del hotels['address.@type']
del hotels['address.addressCountry.@type']
del hotels['geo.@type']
del hotels['starRating.@type']
del hotels['aggregateRating.@type']
del hotels['description']
del hotels['address.addressCountry.name']
```

Преобразуем время в datetime

```python
hotels['checkinTime'] = pd.to_datetime(hotels['checkinTime'], format='%H:%M').dt.time
hotels['checkoutTime'] = pd.to_datetime(hotels['checkoutTime'], format='%H:%M').dt.time
```

Заменяем nan на моду и 'no information'

```python
hotels['address.streetAddress'] = hotels['address.streetAddress'].fillna('No information')
```

```python
hotels['amenityFeature'] = hotels['amenityFeature'].fillna('No information')
```

```python
hotels['checkinTime'].isna().sum()
```

```python
hotels['checkinTime'].value_counts()
```

```python
hotels['checkinTime'] = hotels['checkinTime'].fillna('14:00:00')
```

```python
hotels['checkoutTime'] = hotels['checkoutTime'].fillna('12:00:00')
```

```python
hotels.isna().any()
```

Создадим функцию, которая преобразует json в список объектов, чтобы преобразовать столбец  amenityFeature

```python
import ast

def extract_amenities(row):
    try:
        amenities = ast.literal_eval(row)  # Преобразуем строку в список
        return [feature['name'] for feature in amenities if 'name' in feature]
    except (ValueError, SyntaxError):
        return []

hotels['amenityFeature'] = hotels['amenityFeature'].apply(extract_amenities)
```
