```python
# TODO - еще раз проверить границы
russia_bbox = {
    "lon_min": 19.6389,  
    "lat_min": 41.1851,  
    "lon_max": 180.0000, 
    "lat_max": 81.8569,  

    "kinds": "accomodations",
    "format": "geojson",
    "apikey": opentripmap
}
response_bbox_accommodations = requests.get(f"{BASE_URL}/bbox", params=russia_bbox)

data_bbox = response_bbox_accommodations.json()
print(json.dumps(data_bbox, indent=4, ensure_ascii=False))
```

```python

features = data_bbox.get("features", [])
df = pd.DataFrame([
    {
        "id": feature.get("id"),
        "xid": feature["properties"].get("xid"),
        "name": feature["properties"].get("name", "Unnamed"),
        "longitude": feature["geometry"]["coordinates"][0],
        "latitude": feature["geometry"]["coordinates"][1],
        "distance": feature["properties"].get("dist"),
        "rate": feature["properties"].get("rate"),
        "osm": feature["properties"].get("osm"),
        "kinds": feature["properties"].get("kinds"),
    }
    for feature in features
])

df
```
