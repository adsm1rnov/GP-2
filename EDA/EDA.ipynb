# Цена и качество

```python
df_rating = df_pos
```

```python
df_norm = df_rating[df_rating['min_price']>0]
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(x=df_norm["stars"], order=sorted(df_norm["stars"].unique()))
plt.title("Распределение звездности отелей")
plt.xlabel("Количество звезд")
plt.ylabel("Число отелей")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(df_norm["min_price"], bins=50, kde=True)
plt.title("Распределение минимальной цены отелей")
plt.xlabel("Минимальная цена")
plt.ylabel("Число отелей")
plt.show()
```

```python
fig = px.box(
    df_norm,
    x="stars",
    y="min_price",
    title="Распределение цен в зависимости от звездности отеля",
    labels={"stars": "Количество звезд", "min_price": "Минимальная цена"}, color="stars"
)

fig.update_yaxes(type="log")

fig.show()
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(x=df_norm["stars"], y=df_norm["min_price"])
plt.title("Распределение цен в зависимости от звездности отеля")
plt.yscale("log")
plt.xlabel("Количество звезд")
plt.ylabel("Минимальная цена")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(x=df_norm["stars"], y=df_norm["rating"])
plt.title("Распределение рейтинга в зависимости от звездности отеля")
plt.xlabel("Количество звезд")
plt.ylabel("Рейтинг")
plt.show()
```

Существуют дорогие отели, которые получают плохие оценки, и это не редкость.
Они имеют стандартные удобства (нет уникальных фишек)< плохой сервис или несоответствие ожиданиям клиентов. Расположение может не соответствовать цене.


**Бизнес-гипотеза:**
Высокая цена ≠ качество. Клиенты ожидают соответствия цены и сервиса, а дорогие отели с посредственным сервисом получают низкие рейтинги.
